# Cleaning Patients Data From US Client - Step 2

```
import pandas as pd
patientsCSV=pd.read_csv('/content/gdrive/My Drive/Patients.csv', encoding='windows-1254', error_bad_lines=False)
```

```
patientsCSV.info()
```

```
patientsData = patientsCSV[["LastName", "FirstName", "MiddleName", "SSN", "ClientID", "DateOfBirth"]]
print(patientsData)
```

Find all rows where all values are NaN

```
idx = patientsData.index[patientsData.isnull().all(1)]
nans = patientsData.loc[idx]
nans
```

Delete all rows where all values are NaN

```
patientsData.dropna(axis=0, how='all', inplace=True)
patientsData
```

```
patientsData['inactive_Flag'] = 0
```

```
print(patientsData)
```

```
patientsData.isnull().sum()
```

Find out all rows for which FirstName, ClientID & DateOfBirth are null

```
cols = ['FirstName', 'ClientID', 'DateOfBirth']
patientsDataWithCIDOBNull = patientsData[patientsData[cols].isna().all(axis=1)]
print(patientsDataWithCIDOBNull)
```

Update inactive_Flag to 1 if FirstName, ClientID and DateOfBirth all is null

```
def get_active_flag_status(x):
  if x:
    return 1
  else:
    return 0

patientsData["inactive_Flag"] = patientsData[cols].isna().all(axis=1).apply(lambda x: get_active_flag_status(x))
```

```
activePatientsData = patientsData[patientsData['inactive_Flag'] == 0]
activePatientsData
```

```
InActivePatientsData = patientsData[patientsData['inactive_Flag'] == 1]
InActivePatientsData
```

```
DataWithLastNameSorting = activePatientsData.sort_values('LastName')
```

```
DataWithLastNameSorting.to_csv(r'/content/gdrive/My Drive/LastNameList.csv')
```

```
DataWithLastNameSorting[DataWithLastNameSorting['LastName'].str.contains("NO")] 
```

Search multiple substring in a single column

```
pd.set_option('display.max_columns', None)
sub_strings = [' NO','DO NOT USE']  
regexStr = '|'.join(sub_strings)
textData  = DataWithLastNameSorting[DataWithLastNameSorting['LastName'].str.upper().str.contains(regexStr)]
textData.info()
textData[1:50]
```

Search a substring in all the columns of a data frame

```
DataWithLastNameSorting[DataWithLastNameSorting.apply(lambda row: row.astype(str).str.contains('NO').any(), axis=1)]
```
